```python
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
x=np.linspace(40,300,2600)
```

```python
def weight(zpt):
    w=[]
    for i in zpt:
        if (i>=40 and i<80):
            z_w=((0.93-1)/(np.log10(80)-np.log10(40)))*(np.log10(i)-np.log10(40))+1
            w.append(z_w)
        if (i>=80 and i<151):
            z_w=((0.8-0.93)/(np.log10(151)-np.log10(80)))*(np.log10(i)-np.log10(80))+0.93
            w.append(z_w)
        if (i>=151):
            z_w=0.80
            w.append(z_w)
    return w
```

```python
y=weight(x)
```

```python

sns.set_style("ticks")
plt.plot(x,y)
plt.xscale('log')
plt.show()
```
